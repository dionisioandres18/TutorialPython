## Variables

Las variables son contenedores que sirven para almacenar datos y/o valores.

No existe comando para crear una variable. Una variable se crea en el momento en que se le asigna un valor por primera vez.

Ejemplo:

```python
x = 100
nombre = "Michael"
```

Una variable puede tener un nombre corto (ej: x, y, z) o un nombre mas descriptivo (ej: edad, direccion).

Existen reglas para las variables de python:
* Un nombre de variable debe comenzar con una letra o guion bajo
* Un nombre de variable no puede comenzar con un número
* Un nombre de variable solo puede contener caracteres alfanuméricos y guiones bajos (Az, 0-9 y _)
* Los nombres de las variables distinguen entre mayúsculas y minúsculas (edad, Edad y EDAD son tres variables diferentes)
* Un nombre de variable no puede ser ninguna de las palabras clave de Python (las veremos mas adelante).

Ejemplo:

```python
myvar = "Matias"
my_var = "Matias"
_my_var = "Matias"
myVar = "Matias"
MYVAR = "Matias"
myvar2 = "Matias"
```

```python
# Nombres de variables no validas
my-var = "Batman"
2myvar = "Batman"
my var = "Batman"
```

```python
# Camel Case
# Cada palabra empieza con mayuscula excepto la primera

myVariableName = "Pelusa"

# Pascal Case
# Cada palabra empieza con letra mayuscula

MyVariableName = "Luna"

# Snake Case
# Cada palabra es separada por guion bajo

my_variable_name = "Jupiter"
```

## Tipos de Datos en Python
Las variables pueden almacenar datos de diferentes tipos, y diferentes tipos pueden hacer cosas diferentes.

Python tiene los siguientes tipos de datos integrados de forma predeterminada, en estas categorías:
* Text Type: 	str
* Numeric Types: 	int, float, complex
* Sequence Types: 	list, tuple, range
* Mapping Type: 	dict
* Set Types: 	set, frozenset
* Boolean Type: 	bool
* Binary Types: 	bytes, bytearray, memoryview
* None Type: 	NoneType

Podemos conocer el tipo de dato con el que estamos trabajando usando la funcion ***type()***:


```python
x = 30 # int (entero)
y = 3.5 # float (coma flotante o decimal)
z = 3+4j # complex (complejo)
nombre = "Anibal" # str (string:cadena de texto)
ciudades = ["Bogotá", "Cartagena", "Barranquilla", "Medellin"] # list (lista)
automovil = {"marca" : "Chevrolet", "modelo" : "Tracker", "color" : "Negro", "año": 2023} # dict (diccionario)
print(type(x))
print(type(y))
print(type(z))
print(type(nombre))
print(type(ciudades))
print(type(automovil))
```

## Conversión de tipo (Numeros)
Puede convertir de un tipo a otro con los métodos int(), float()y :complex()


Ejemplos:

```python
x = 1    # entero
y = 2.8  # flotante
z = 1j   # complejo

#convertir de entero a flotante:
a = float(x) # ----> a=1.0

#convertir de flotante a entero:
b = int(y) # ----> b=2

#convertir de entero a complejo:
c = complex(x) # ----> c=1+0j

print("a:",a,"---->",type(a))
print("b:",b,"---->",type(b))
print("c:",c,"---->",type(c))
```

## Conversión del tipo de dato
la conversión en python se realiza mediante funciones de constructor:

  * int() : construye un número entero a partir de un literal entero, un literal flotante (eliminando todos los decimales) o un literal de cadena (siempre que la cadena represente un número entero)
  * float() : construye un número flotante a partir de un literal entero, un literal flotante o un literal de cadena (siempre que la cadena represente un flotante o un entero)
  * str() : construye una cadena a partir de una amplia variedad de tipos de datos, incluidas cadenas, literales enteros y literales flotantes

  Ejemplo:


```python
# Conversion a enteros
x = int(1)   # x will be 1
y = int(2.8) # y will be 2
z = int("3") # z will be 3

print(x, type(x))
print(y, type(y))
print(z, type(z))
```

```python
# Conversin a flotantes
x = float(1)     # x will be 1.0
y = float(2.8)   # y will be 2.8
z = float("3")   # z will be 3.0
w = float("4.2") # w will be 4.2

print(x, type(x))
print(y, type(y))
print(z, type(z))
```

```python
#Conversion a cadenas o strings
x = str("s1") # x will be 's1'
y = str(2)    # y will be '2'
z = str(3.0)  # z will be '3.0'

print(x, type(x))
print(y, type(y))
print(z, type(z))
```

## Slicing Strings
Las cadenas pueden rebanarse, es decir, se puede extraer una parte de ellas si se quiere. Para extraer una parte de una cadena es necesario especificar el índice inicial y el índice final, separados por dos punto. *(esta extracción incluye el valor del indice inicial, pero no el inidice final)*

Ejemplo:

```python
# palabra = P y t h o n
# indice  = 0 1 2 3 4 5
# indice inicial = 0 --> P
# indice final = 4 --> o pero como no lo incluye, se toma el indice anterior, es decir el 3 --> h
palabra = "Python"
palabra[0:4]
```

```python
# estacion = p r i m a v e r a
# indice   = 0 1 2 3 4 5 6 7 8
# indice inicial = 3 --> m
# indice final = 7 --> r pero como no lo incluye, se toma el indice anterior, es decir el 6 --> e
estacion = "primavera"
estacion[3:7]
```

```python
# Al omitir el índice de inicio, el rango comenzará en el primer carácter:
pais = "Colombia"
pais[:5]
```

```python
# Al omitir el índice final , el rango irá hasta el final:
pais = "Colombia"
pais[3:]
```

```python
# Utilice índices negativos para iniciar el segmento desde el final de la cadena:
# pais            =  C  o  l  o  m  b  i  a
# indice negativo = -8 -7 -6 -5 -4 -3 -2 -1
pais = "Colombia"
pais[-4]
```

```python
# Las cadenas no son editables
word="bananas"
word[0]="B"
```

## Concatenación de strings
Para concatenar o combinar dos cadenas, puede usar el operador +.

Ejemplo:

```python
nombre = "Luisa"
apellido = "Vargas"

print(nombre+apellido)
print(nombre+" "+apellido) # Añadiendo un espacio
```

Podemos combinar numeros y cadenas usando el método format().

El método format() toma los argumentos que esten dentro del parentesis, y los coloca en la cadena donde están los marcadores de posición {}.

Ejemplo:

```python
nota_final = 3.6
texto = "Samuel obtuvo una calificacion de {} en su examen de inglés"
print(texto.format(nota_final))
```

```python
altura = 1.72
edad = 21
texto = "Fernanda tiene {} años y mide {} metros"
print(texto.format(edad, altura))
```
