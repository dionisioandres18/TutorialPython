
# **Condicionales**
Son sentencias que te permiten condicionar una decisión entre la elección de una opción y otra. En nuestra vida cotidiana usamos muchos condicionales por ejemplo: "si el semaforo esta en rojo, detengo el auto"

La instruccion ***if*** es usada para ejecutar  determinadas instrucciones en bloque si se cumple una determinada condicion representada por una expresion booleana(verdadera o falsa). Existen varias formas en la auqe que una isntruccion condicional puede aparecer:

1. if simple
2. Varios if simple anidados
3. if-else
4. Varios if-else aninados
5. if-elif-else

Python admite operaciones lógicas habituales
* Igualdad: a == b
* Diferente de: a != b
* Menor que: a < b
* Menor o igual a: a <= b
* Mayor que: a > b
* Mayor o igual a: a >= b

Son operaciones que suelen ser utilizadas comunmente en declaraciones "if" y bucles.





## if Simple
Ejecutará un bloque de instrucciones si una determinada expresion logica es verdadera y no ejecutará nada si es falsa.

```python
#
# Ejemplo: Programa que muestra si un numero es mayor a otro
#

a = 33
b = 200
if (b > a): #si b es mayor que a imprima "b is greater than a"
  print("b is greater than a")
```

```python
#
# Ejemplo: Si a es mayor a 7 imprima verdadero
#

a=8
if a>7:
  print("verdadero")
```

```python
#
# Ejemplo: Si a es mayor a 7 imprima verdadero
#

a=5
if a>7: # en este ejemplo a es menor que 7 por lo tanto la expresion es falsa y el codigo no ejecutará nada
  print("verdadero")
```

## Varios if-simple aninados
Es posible colocar un if simple dentro del bloque de instrucciones de otro if con el fin de evaluar diferentes condiciones. Podemos poner tantos if aninados como queramos. Ejemplo:

```python
#
# Queremos saber si un numero a, esta entre 10 y 20 o si  es mayor a 20
#

a=21
if a>10: #a es mayor a 10 por tanto es verdadero y se ejecuta el bloque esta dentro del if
  if a<=20: # a no es menor o igual a 20, por lo tanto no se ejecuta el bloque dentro del if
    print(a," esta entre 10 y 20")
  if a>20: # a es mayor a 20 por lo tanto  se ejecuta el bloque de instrucciones dentro
    print(a," es mayor que 20") # se muestra en pantalla este mensaje
```

```python
#
# Queremos saber si un numero a, esta entre 10 y 20 o si  es mayor a 20
#

a=17
if a>10: #a es mayor a 10 por tanto es verdadero y se ejecuta el bloque esta dentro del if
  if a<=20: # a es menor o igual a 20, por lo tanto se ejecuta el bloque dentro del if
    print(a," esta entre 10 y 20") # se muestra en pantalla este mensaje
  if a>20: # a no es mayor a 20 por lo tanto no se ejecuta el bloque de instrucciones dentro
    print(a," es mayor que 20") # no se muestra en pantalla este mensaje
```

## if-else
La instrucción if-else añada al if cimple la posibilidad de ejecutar un bloque de instrucciones cuando la condicion es falsa. Ejemplo

```python
#
# Progrma que muestra si un numero es negativo o positivo (en este ejemplo no se incluira el cero)
#

# Pedimos al usuario que ingrese un numero usando la funcion input
numero = int(input("Ingrese un numero: "))

if numero>0: #si el numero es mayor que cero, se ejecuta la instruccion dentro del if
  print(numero, " es positivo")
else: #si el numero no es mayor que cero, se ejecuta la instruccion dentro del else
  print(numero," es negativo")
```
